```python
import bs4
from newspaper import Article
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen



#GOOGLE_NEWS_URL = 'https://www.google.co.jp/search?q={}&hl=ja&source=lnt&tbs=cdr%3A1%2Ccd_min%3A{}%2Ccd_max%3A{}&tbm=nws&start={}'

#news_url="https://news.google.com/news/rss"
keyword = 'Ikea'
news_url = "https://news.google.com/news?pz=1&cf=all&q=" + keyword + "&tbs=cdr%3A1%2Ccd_min%3A4%2F6%2F2017%2Ccd_max%3A1%2F12%2F2018&start=1" +"&output=rss"
#news_url = 'https://news.google.com/news/rss/search/section/q/Ikea?tbs=cdr%3A1%2Ccd_min%3A4%2F6%2F2017%2Ccd_max%3A1%2F12%2F2018'
Client=urlopen(news_url)
xml_page=Client.read()
Client.close()

soup_page=soup(xml_page,"xml")
news_list=soup_page.findAll("item")
# Print news title, url and publish date
for news in news_list:
    #print(news.title.text)
    #print(news.link.text)
    print(news.pubDate.text)
    
    
    url = news.link.text
    article = Article(url)
    article.download()
    
    article.parse()
    #print(article.authors)
    #print(article.text)
    print(article.title)
    
    article.nlp()
    print(article.keywords)
    print(article.summary)
    print("-"*60)

```
